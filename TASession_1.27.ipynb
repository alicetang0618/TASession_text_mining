```python
from nltk.corpus import words
words.words()
```

```python
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unusual = text_vocab - english_vocab
    return sorted(unusual)
```

```python
unusual_words(text1)
```

```python
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return len(content) / len(text)
```

```python
content_fraction(text1)
```

## 2. Pronouncing Dictionary

```python
# (word, [phonetic codes])
# pay attention to the words with multiple entries
entries = nltk.corpus.cmudict.entries()
entries[:100]
```

```python
for entry in entries:
    word = entry[0]
    pron = entry[1]
    if len(pron) == 3:
        ph1, ph2, ph3 = pron
        if ph1 == 'P' and ph3 == 'T':
            print(word, ph2, end=' ')
```

```python
# note: vowel sounds end with a digit
syllable = ['N', 'IH0', 'K', 'S']
[word for word, pron in entries if pron[-4:] == syllable]
```

## 3. WordNet

```python
# wn: a tree structure of similar words
# synset: a set of similar words
from nltk.corpus import wordnet as wn
wn.synsets('car')
```

```python
wn.lemmas('car')
```

```python
# a synset contains lemmas
wn.synset('car.n.01').lemmas()
```

```python
wn.lemma('car.n.01.car').synset()
```

```python
# hyponyms and hypernyms
motorcar = wn.synset('car.n.01')
types_of_motorcar = motorcar.hyponyms()
types_of_motorcar
```

```python
motorcar.hypernyms()
```

```python
paths = motorcar.hypernym_paths()
paths[1]
```

```python
motorcar.min_depth()
```

```python
# semantic similarity of two synsets
right = wn.synset('right_whale.n.01')
orca = wn.synset('orca.n.01')
minke = wn.synset('minke_whale.n.01')
tortoise = wn.synset('tortoise.n.01')
novel = wn.synset('novel.n.01')
right.lowest_common_hypernyms(minke)[0].min_depth()
```
